```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame

study_data_complete = pd.merge(study_results, mouse_metadata, how ="left", on ="Mouse ID")
# Display the data table for preview
study_data_complete.head(10)
```

```python
# Checking the number of mice.
# By using unique() abd len() functions extract an arry of uniqu values form "Mouse ID" columns 
# and use len() to calculate the length of the array representing the total count of the unique mice. 
mouse_count = len(study_data_complete["Mouse ID"].unique())
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
# The .loc function is used to filter the rows of studystudy_data_complete
duplicate_mouse_ids = study_data_complete.loc[study_data_complete.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()

duplicate_mouse_ids[0]
```

```python
# Optional: Get all the data for the duplicate mouse ID.
# Extract all rows from the study_data_complete DataFrame where the Mouse ID matches the first entry in the duplicate_mouse_ids array.
# The .loc function is used to filter the rows of studystudy_data_complete
duplicate_mouse_data = study_data_complete.loc[study_data_complete["Mouse ID"]== duplicate_mouse_ids[0]]
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# for this specific task we can also uses drop_duplicates() function 
clean_data = study_data_complete[study_data_complete["Mouse ID"].isin(duplicate_mouse_ids)==False]
clean_data.head() # Return the fisrt 5 raws of the dataframe 
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice_data_count = len(clean_data["Mouse ID"].unique())
clean_mice_data_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# use groupby function to group the rows of clean_data by unique values in the column "Drug Regimen"
# and selsct the column "Tumore Volume (mm3)" from the grouped data to aggrigate calculations of the tumor volume for each drug. 
grouped_drug_data = clean_data.groupby("Drug Regimen")["Tumor Volume (mm3)"]

# Calculate summary statistics
means = grouped_drug_data.mean() # this calculate the average of tumor volume for each drug regimen 
medians = grouped_drug_data.median() # this calculate the middle value tumor Volume for each drug regimen group 
variances = grouped_drug_data.var() # this calculate the variance of the Tumor Volume for each drug regimen group 
std_deviation = grouped_drug_data.std() # by using .std, it calculate the standard deviation of tumor volume for each drug regimen
sems = grouped_drug_data.sem() # by using .sem, it calculates the sem of tumor volume for each drug regimen 

# Assemble the resulting series into a single summary DataFrame
summary_stat = pd.DataFrame({
    "Mean Tumor Volume": means,
    "Median Tumor Volume": medians,
    "Tumor Volume Variance": variances,
    "Tumor Volume Std. Dev.": std_deviation,
    "Tumor Volume Std. Err.": sems
})

# Display the resulting summary statistics DataFrame
df= summary_stat
df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
# Grouped the unique data placed in Drug Regimen column
# we use .agg() function to apply aggrigation on the grouped data
df1 = clean_data.groupby("Drug Regimen").agg(
    {"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]} # this is a dictionary to apply .agg() int the Tumor volume column 
)
df1
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

# REMINDER: clean_data is a product of to marged dataset ie Mouse_metadata and Study_results
maxTime = clean_data.groupby("Mouse ID")["Timepoint"].max() # Group the unique value in the "Mouse Id " column and select timepoint

# NOTE: the series needs to be converted to a dataframe to be merged with another dataframe. 
maxTime = maxTime.reset_index() # convert the maxTime series in to a data frame to make it ready from merge

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_data = maxTime.merge(clean_data, on=["Mouse ID", "Timepoint"], how="left")
merged_data.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"] # this is a list that contain the name of the drug


# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list = [] # an empty list that will contain the tumer vol data latter on 


# Calculate the IQR and quantitatively determine if there are any potential outliers.
for treatment in treatment_list: # Iterate over each element in the treatment_list (each treatment regemin)

    # Locate the rows which contain mice on each drug and get the tumor volumes
    # selects the rows where the âDrug Regimenâ column matches the current treatment in the loop 
    last_tumor_vol_data = merged_data.loc[merged_data["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
   
    # Append the tumor_volume_list that contains the tumor volume for the current treatment
    #regimen to the the tumor_volume_list 
    tumor_vol_list.append(last_tumor_vol_data)
    
    # Calculate the 25th percentile, 50th percentile (median), and 75th percentile of the tumor volume data.
    quartiles = last_tumor_vol_data.quantile([.25, .5, .75])

    # Extract lower and upper values fron the quartiles series 
    lowerq = quartiles [0.25]
    upperq = quartiles [0.75]

    # Calculate the Interquartile Range (IQR), which is the range of the middle 50% of the data.
    IQR = upperq - lowerq

    # Determine outliers using upper and lower bounds 
    lower_bound = lowerq - (1.5*IQR)
    upper_bound = upperq + (1.5*IQR)

    # Aplly logical condition fo filter out data point in last_tumor_vol_data
    # Outlier for each drug regimen will be idenfified 
    outliers = last_tumor_vol_data.loc[(last_tumor_vol_data < lower_bound) | (last_tumor_vol_data > upper_bound)]
    
    if len(outliers) == 0:  
        # If no outliers, print a message indicating no potential outliers for the current treatment
        print(f"{treatment}'s potential outliers: None")
    else:  
        # If outliers are found, initialize the output message with the treatment name
        output = f"{treatment}'s potential outliers:"
        # Loop through each outlier value
        for i in outliers:  
            # Append the outlier value to the output message
            output += f" {i}"
        # Print the final message listing the outliers
        print(output)
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.

plt.boxplot(tumor_vol_list, 
            notch=False,  # Disable notch for the box plot
            patch_artist=True,  # Fill the boxes with color
            boxprops=dict(facecolor='skyblue', color='black'),  # Set box color
            whiskerprops=dict(color='black', linewidth=1),  # Set whisker color and linewidth
            capprops=dict(color='black'),  # Set cap color
            flierprops=dict(markerfacecolor='red', marker='o', markersize=10),  # Set outlier properties
            medianprops=dict(color='black', linewidth=2))  # Set median line color and width
            

# Add labels and title
plt.title("The distribution of the tumor volume for each treatment group.")
plt.xlabel("Treatment Regimen")
plt.ylabel("Tumor Volume (mm3)")

# Set x-ticks to the treatment names
plt.xticks(ticks=range(len(treatment_list)), labels=treatment_list)

# Show the plot
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(capomulin_average["Weight (g)"],
                          capomulin_average["Tumor Volume (mm3)"])[0]


# Define x and y values
x_values = capomulin_average["Weight (g)"]
y_values = capomulin_average["Tumor Volume (mm3)"]

# Perform linear regression
# Perform a linear regression analysis on the provided x and y values.
# The linregress function from the scipy.stats module calculates the following:
(slop, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)

# 'slop' (slope): The rate of change of y with respect to x (m in y = mx + b).
# 'intercept': The value of y when x = 0 (b in y = mx + b).
# 'rvalue': The correlation coefficient (r), indicating the strength and direction of the linear relationship.
# 'pvalue': The probability of observing the data assuming the null hypothesis (no relationship between x and y).
# 'stderr': The standard error of the slope, indicating the variability in the slope estimate.

# Calculate the y-values for the linear regression line using the equation y = mx + b
# 'slop' is the slope, 'x_values' is the independent variable, and 'intercept' is the y-intercept.
linear_regr = x_values * slop + intercept  

# Create a string representation of the linear equation in the format: "y = mx + b"
# The slope and intercept are formatted to 2 decimal places for readability.
lineEq = f"y = {slop: .2f}x + {intercept:.2f}"

# Create a scatter plot for the data points
plt.scatter(x_values, y_values)
# The "b" specifies the line color red and style (solid).
plt.plot (x_values, linear_regr, "b-")

plt.xlabel ("Weight (g)")
plt.ylabel ("Average Tumor Volume (mm3)")
plt.title ("Average Tumor Vs. Mouse Weight for the Campomulin Regimen")
print(f"The correlation between mouse weight and the average tumor volume is {correlation:.2f}")
plt.show()
```
